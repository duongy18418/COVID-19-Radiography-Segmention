```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Segformer Model

### Evaluation Metrics

```python
seg_metrics = pd.read_csv("./results/covid_segformer.csv")

iou = seg_metrics['IoU'].to_list()
accuracy = seg_metrics['Accuracy'].to_list()
precision = seg_metrics['Precision'].to_list()
recall = seg_metrics['Recall'].to_list()
f1 = seg_metrics['F1'].to_list()

iou.sort()
accuracy.sort()
precision.sort()
recall.sort()
f1.sort()

print('Segformer Model Evaluation Metrics')

plt.figure(figsize=(20,10))

ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)

ax1.plot(iou)
ax2.plot(accuracy)
ax3.plot(precision)
ax4.plot(recall)
ax5.plot(f1)

ax1.set_title('IoU Score')
ax2.set_title('Accuracy Score')
ax3.set_title('Precision Score')
ax4.set_title('Recall Rate')
ax5.set_title('F1 Score')
```
